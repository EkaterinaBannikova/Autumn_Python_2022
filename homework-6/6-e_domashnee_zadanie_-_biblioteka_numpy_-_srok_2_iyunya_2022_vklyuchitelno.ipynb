7. –ü—Ä–∏–º–µ–Ω–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é dot –¥–ª—è –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä –Ω–∞ –≤–µ–∫—Ç–æ—Ä, –º–∞—Ç—Ä–∏—Ü—É –Ω–∞ –≤–µ–∫—Ç–æ—Ä –∏ –º–∞—Ç—Ä–∏—Ü—É –Ω–∞ –º–∞—Ç—Ä–∏—Ü—É.

```python
Vector1 = np.array([1, 2, 3])
Vector2 = np.array([4, 5, 6])
x = np.dot(Vector1,Vector2)

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
y = np.dot(A, B)

C = np.array([ [1, 2], [3, 4], [5, 6] ])
Vector3 = np.array([7, 8])
z = np.dot(C, Vector3)

print('Vector1 x Vector2 =',x)
print('A x B =',y)
print('C x Vector3 =',z)
```

8. –í—ã—á–∏—Å–ª–∏—Ç–µ cos(x) –∏ sin(x) –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ [0,1] —Å —à–∞–≥–æ–º $0.05$, –∞ –∑–∞—Ç–µ–º –æ–±—ä–µ–¥–∏–Ω–∏—Ç–µ –æ–±–∞ –º–∞—Å—Å–∏–≤–∞ —á–∏—Å–µ–ª –∫–∞–∫ —Å—Ç—Ä–æ–∫–∏ –≤ –æ–¥–∏–Ω –º–∞—Å—Å–∏–≤.

```python
x = np.linspace(0,1,21)
y = np.cos(x)
z = np.sin(x)
a = np.vstack((y, z)) 
print('cos=',y)
print('sin=',z)
print(a)
```

10. –ù–∞–π–¥–∏—Ç–µ —Å—É–º–º—É –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤, —Å—É–º–º—É –≤–Ω—É—Ç—Ä–∏ —Å—Ç—Ä–æ–∫, —Å—É–º–º—É –≤–Ω—É—Ç—Ä–∏ —Å—Ç–æ–ª–±—Ü–æ–≤, –∞ —Ç–∞–∫–∂–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —á–∏—Å–µ–ª –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã $A$. (–ü–æ–¥–æ–±–Ω–æ —Ç–æ–º—É, –∫–∞–∫ sorted –∏–º–µ–µ—Ç –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç key=, –º–Ω–æ–≥–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ Numpy –∏–º–µ—é—Ç –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç axis=)


```python
a = np.array([[1, 5, 3, 4, 7],[6, 2, 9, 1, 6],[1, 1, 8, 4, 6],[2, 7, 7, 1, 9]])
print('–°—É–º–º–∞ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ = ',np.sum(a))
print('–°—É–º–º–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö = ',np.sum(a, axis = 0))
print('–°—É–º–º–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫–∞—Ö = ',np.sum(a, axis = 1))
print('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ = ',np.mean(a, axis=1))
print('–î–∏—Å–ø–µ—Ä—Å–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ = ',np.var(a, axis=1))
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ = ', np.sqrt(np.var(a, axis = 1)))
```

11.  –û—Ç–Ω–∏–º–∏—Ç–µ –æ—Ç –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Ç—Ä–∏—Ü—ã $A$ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ —Å—Ç—Ä–æ–∫–µ –∏ –ø–æ–¥–µ–ª–∏—Ç–µ –Ω–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏.

```python
x = np.average(a, axis=1)
y = np.sqrt(np.var(a, axis=-1, dtype=float))
print((a - x.reshape(len(x), 1)) / y.reshape(len(y), 1))
```

13. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –¥–≤–µ –º–∞—Ç—Ä–∏—Ü—ã: $B = A^TA$ –∏ $C = AA^T$. –ß—Ç–æ –≤—ã –º–æ–∂–µ—Ç–µ –æ –Ω–∏—Ö —Å–∫–∞–∑–∞—Ç—å? –ö–∞–∫–æ–≥–æ –æ–Ω–∏ —Ä–∞–∑–º–µ—Ä–∞? –í —á—ë–º –∏—Ö —Å—Ö–æ–¥—Å—Ç–≤–æ –∏ –æ—Ç–ª–∏—á–∏–µ?

14. –ù–∞–π–¥–∏—Ç–µ —Å—É–º–º—É –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü $B$ –∏ $C$.

```python
x = np.trace(b)
y = np.trace(c)
print('–°—É–º–º–∞ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü  ùêµ =', x)
print('–°—É–º–º–∞ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü  C =', y)
```

(–ø–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≥–æ—Ç–æ–≤—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ np.linalg). 


```python
x = np.linalg.matrix_rank(a)
y = np.linalg.matrix_rank(b)
z = np.linalg.det(b)
print('–†–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã –ê = ',x,'\n–î–µ—Ç–µ—Ä–º–∏–Ω–∞–Ω—Ç –∏ —Å–ø–µ–∫—Ç—Ä –º–∞—Ç—Ä–∏—Ü—ã A –ø–æ—Å—á–∏—Ç–∞—Ç—å –Ω–µ–ª—å–∑—è,—Ç.–∫. –º–∞—Ç—Ä–∏—Ü–∞ –Ω–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è ')
print('–†–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã –í = ',y,'\n–î–µ—Ç–µ—Ä–º–∏–Ω–∞–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ã –í = ',z)
w, v = np.linalg.eig(b)
f = np.linalg.matrix_rank(c)
e = np.linalg.det(c)
print ("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –º–∞—Ç—Ä–∏—Ü—ã –í = :\n", w)
print ("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã –º–∞—Ç—Ä–∏—Ü—ã –í = :\n", v)
print('–†–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã C = ',f,'\n–î–µ—Ç–µ—Ä–º–∏–Ω–∞–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ã C = ',e)
w, v = np.linalg.eig(c)
print ("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –º–∞—Ç—Ä–∏—Ü—ã C = :", w)
print ("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã –º–∞—Ç—Ä–∏—Ü—ã C = :\n", v)
```

 $\int\limits_0^2 {\frac{{2x - 1}}{{2x + 1}}dx}=\int\limits_0^2 {(1-\frac{{2}}{{2x + 1}})dx}=\int\limits_0^2 {1dx}-2\int\limits_0^2 {(\frac{{1}}{{2x + 1}})dx}=(t=2x+1)=2-2\int\limits_0^2 {(\frac{{1}}{{2t}})dt}=2-1\int\limits_0^2 {(\frac{{1}}{{t}})dt}=2-ln|4+1|=2-ln5=0.39056$

```python
def f(x):
    return (2 * x - 1) / (2 * x + 1)

def method_rectangle(f, a, b, step):
    x = np.linspace(a, b - step, int((b - a) / step))
    y = f(x) * step
    return np.sum(y)

def method_trapezoid(f, a, b, step):
    x = np.linspace(a, b, int((b - a) / step) + 1)
    y = f(x[0]) + 2 * np.sum(f(x[1:-1])) + f(x[-1])
    return y * step / 2

def method_simpson(f, a, b, step):
    x = np.linspace(a, b, int((b - a) / step) + 1)
    y = f(x[0]) + 4 * np.sum(f(x[1:-1:2])) + 2 * np.sum(f(x[2:-1:2])) + f(x[-1])
    return y * step / 3

print('–ú–µ—Ç–æ–¥ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤:', method_rectangle(f, 0, 2, 0.00001))
print('–ú–µ—Ç–æ–¥ —Ç—Ä–∞–ø–µ—Ü–∏–π:', method_trapezoid(f, 0, 2, 0.001))
print('–ú–µ—Ç–æ–¥ –°–∏–º–ø—Å–æ–Ω–∞:', method_simpson(f, 0, 2, 0.01))
```
