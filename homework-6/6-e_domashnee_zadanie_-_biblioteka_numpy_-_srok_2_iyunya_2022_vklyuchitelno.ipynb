 $\int\limits_0^2 {\frac{{2x - 1}}{{2x + 1}}dx}=\int\limits_0^2 {(1-\frac{{2}}{{2x + 1}})dx}=\int\limits_0^2 {1dx}-2\int\limits_0^2 {(\frac{{1}}{{2x + 1}})dx}=(t=2x+1)=2-2\int\limits_0^2 {(\frac{{1}}{{2t}})dt}=2-1\int\limits_0^2 {(\frac{{1}}{{t}})dt}=2-ln|4+1|=2-ln5=0.39056$

```python
def f(x):
    return (2 * x - 1) / (2 * x + 1)

def method_rectangle(f, a, b, step):
    x = np.linspace(a, b - step, int((b - a) / step))
    y = f(x) * step
    return np.sum(y)

def method_trapezoid(f, a, b, step):
    x = np.linspace(a, b, int((b - a) / step) + 1)
    y = f(x[0]) + 2 * np.sum(f(x[1:-1])) + f(x[-1])
    return y * step / 2

def method_simpson(f, a, b, step):
    x = np.linspace(a, b, int((b - a) / step) + 1)
    y = f(x[0]) + 4 * np.sum(f(x[1:-1:2])) + 2 * np.sum(f(x[2:-1:2])) + f(x[-1])
    return y * step / 3

print('Метод прямоугольников:', method_rectangle(f, 0, 2, 0.00001))
print('Метод трапеций:', method_trapezoid(f, 0, 2, 0.001))
print('Метод Симпсона:', method_simpson(f, 0, 2, 0.01))
```
